```
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```
A = 10
t = 0.5
x = np.arange(0,4,0.01)
y = A*np.sin(4*x*pi)
```

```
plt.plot(x,y,'--')
plt.legend(['sin'])
plt.title('Grafik Sinus')
plt.ylabel('Amplitudo')
plt.xlabel('t')
plt.show()
```
